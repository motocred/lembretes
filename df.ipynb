### Formating the sellings sheet

```python
def format_df(df_raw: pd.DataFrame, expected_cols: list[str]) -> pd.DataFrame:
    for row_idx in range(len(df_raw)):
        row = df_raw.iloc[row_idx]
        for col_idx in range(len(row)):
            sub_row = row.iloc[col_idx:]
            if list(sub_row.values[:len(expected_cols)]) == expected_cols:
                # Slice out the relevant part of the dataframe
                df_trimmed = df_raw.iloc[row_idx + 1:, col_idx:col_idx + len(expected_cols)]
                df_trimmed.columns = expected_cols
                df_trimmed.reset_index(drop=True, inplace=True)
                return df_trimmed

    raise ValueError("Expected column names not found in the sheet.")
```

### Extract the data necessary

```python
# Declaring dictionary
customers_reminders = {
    "atras": [], # Audio 3 dias
    "closes": [] # Audio 10 dias # Text no dia de vencimento e 1 dia antes
}

customers_pos_sell = { 
    "pos_sell": [] # Uma semana pós venda (audio)
}

customers_execution_error = list()
```

```python
dfs = {
    "wallet": wallet_df,
    "sellings": selling_df_formated,
    "parcs": parcs_df_formated,
    "info_op": info_op_df,
    "atras": atras_df_formated
}
```

```python
def get_sell_code(df_sellings: pd.DataFrame, line: int) -> str:
    return df_sellings["COD_VENDA"][line]

get_sell_code(dfs["sellings"], 9)
```

```python
def get_name_by_code(df_sellings: pd.DataFrame, sell_code: str) -> str:
    return df_sellings.loc[df_sellings["COD_VENDA"] == sell_code, "NOM_CLIENTE"].values[0]

get_name_by_code(dfs["sellings"], '39303148')
```

```python
def get_number_by_name(df_info_op: pd.DataFrame, name: str) -> str:
    df_info_op["CLIENTE"] = df_info_op["CLIENTE"].str.strip().str.upper()
    match_row = df_info_op.loc[df_info_op["CLIENTE"] == name.upper().strip(), "TELEFONE"]

    if not match_row.empty:
        return match_row.values[0]
    
    return None

get_number_by_name(dfs["info_op"], 'LUIS CARLOS BARBOSA DO NASCIMENTO')
```

```python
def get_days_atras(df_atras: pd.DataFrame, line: int) -> int:
    if str(df_atras["VERIF"][line]).upper().strip() == "ATRASADO":
        return int(df_atras["Dias de atraso"][line])

    return 0
```

```python
def get_next_venc_by_sell_code(df_parcs: pd.DataFrame, sell_code: str) -> str:
    df_parcs_sell_code = df_parcs[df_parcs["COD_VENDA"] == sell_code].copy()
    df_parcs_sell_code["DATA_PARC"] = pd.to_datetime(df_parcs_sell_code["DATA_PARC"].str.strip(), format="%d/%m/%y")
    df_parcs_sell_code.sort_values(by="DATA_PARC", inplace=True)

    for _, row in df_parcs_sell_code.iterrows():
        if str(row["VERIF"]).upper().strip() == "NO PRAZO":
            return row["DATA_PARC"].date()

    return None

get_next_venc_by_sell_code(dfs["parcs"], '39303148')
```

```python
def seperate_customers_reminders(customers_reminders: dict[str, list[dict[str, str]]], dfs: dict[str, pd.DataFrame]) -> None:
    today_date = dt.date.today()

    df_sellings = dfs["sellings"]
    df_parcs = dfs["parcs"]
    df_info_op = dfs["info_op"]
    df_atras = dfs["atras"]

    for line in range(len(df_sellings)):
        # Get nex to venc
        try:
            sell_code = get_sell_code(df_sellings, line)
            date_venc = get_next_venc_by_sell_code(df_parcs, sell_code)
            if not date_venc:
                continue

            name = get_name_by_code(df_sellings, sell_code)
            number = get_number_by_name(df_info_op, name)
            days_diff = (date_venc - today_date).days
            
            if days_diff in [10, 1, 0]: # Add to closes
                close = {
                    "name": name,
                    "number": number,
                    "days_to": days_diff,
                    "sell_code": sell_code
                }
                customers_reminders["closes"].append(close)
        except Exception as e:
            customer_error = {
                "name": name,
                "sell_code": sell_code,
                "error": e,
                "type": "Reminder"
            }
            customers_execution_error.append(customer_error)

    for _, row in df_atras.iterrows():
        try:
            if pd.isna(row["Dias de atraso"]):
                continue
            days = int(row["Dias de atraso"])
            name = str(row["NOME"]).strip()
            sell_code = str(row["COD_VENDA"]).strip()
            if days > 0 and days <= 3: # Add to atra
                atra = {
                    "name": name,
                    "number": number,
                    "days_after": days,
                    "sell_code": sell_code,
                }
                customers_reminders["atras"].append(atra)
        except Exception as e:
            customer_error = {
                "name": name,
                "sell_code": sell_code,
                "error": e,
                "type": "Atras"
            }
            customers_execution_error.append(customer_error)
```

```python
seperate_customers_reminders(customers_reminders, dfs)
```

### POS SELLS

```python
def get_pos_sell(dfs: pd.DataFrame) -> list[dict[str, str]]:
    customers_pos_sell = list()
    today = dt.date.today()

    df_wallet = dfs["wallet"]
    df_info_op = dfs["info_op"]

    df_wallet["Data"] = pd.to_datetime(df_wallet["Data"], format="%d-%m-%Y")

    try: 
        for _, row in df_wallet.iterrows():
            date_sell = row["Data"].date()
            if pd.isna(date_sell):
                continue
            days = (today - date_sell).days
            print(days)
            if days == 7:
                name = row["Cliente"]
                number = get_number_by_name(df_info_op, name)
                sell_code = row["Operação BMP"]

                data = {
                    "name": name,
                    "number": number,
                    "sell_code": sell_code
                }

                customers_pos_sell.append(data)
    except Exception as e:
        customer_error = {
            "name": name,
            "sell_code": sell_code,
            "error": e,
            "type": "Pos"
        }
        customers_execution_error.append(customer_error)

    return customers_pos_sell
```
